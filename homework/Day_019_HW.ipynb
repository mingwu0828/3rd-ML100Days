```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件

# 忽略警告訊息
%matplotlib inline
plt.style.use('ggplot')
import warnings
warnings.filterwarnings('ignore')

# 設定 data_path
dir_data = './data/'
```

```python
# 讀取檔案
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

```python
amt_data = app_train[['HOUSETYPE_MODE', 'AMT_CREDIT']]
amt_data['MAMT_BINNED'] = pd.cut(amt_data['AMT_CREDIT'], 
                                  bins = np.linspace(0, 5000000, num = 11))
print(amt_data['MAMT_BINNED'].value_counts())
amt_data.head()
```

```python
amt_group_sorted = np.sort(amt_data['MAMT_BINNED'].unique())
amt_data.head()
```

```python
unique_MAMT_BINNED = amt_data['MAMT_BINNED'].unique()
print(unique_MAMT_BINNED)
```

```python
unique_house_type = amt_data['HOUSETYPE_MODE'].unique()
print(unique_house_type)
```

```python
# 使用不同的 HOUSETYPE_MODE 類別繪製圖形, 並使用 subplot 排版
"""
Your Code Here
"""
unique_house_type = amt_data['HOUSETYPE_MODE'].unique()

nrows = len(amt_group_sorted)
ncols = 2

plt.figure(figsize=(10,30))
for i in range(len(amt_group_sorted)):
    plt.subplot(nrows, ncols, i+1)
    """
    Your Code Here
    """
    for j in range(len(unique_house_type)):
        sns.distplot(app_train.loc[(amt_data['MAMT_BINNED'] == amt_group_sorted[i]) & \
                                   (amt_data['HOUSETYPE_MODE'] == unique_house_type[j]), 'AMT_CREDIT'], 
                     label = unique_house_type[j], hist = False)
        
    plt.title(str(amt_group_sorted[i]))
plt.show()    
```
